## Setup Virtual Server

## **1. Buy a virtual server**

Example:

- Memory: 8 GB
- CPU: 4 Cores
- Hard Disk: 60 GB SSD

## **2. Access Server Using SSH & VSCode**

### 1. Generate SSH key on your local machine

### 2. Connect to virtual server using VSCode

1. Use `Remote Development` extension
2. Select Connect to Host
3. Enter host: *yourusername@ipaddress*
4. Eneter password

### 3. Add public SSH of your local machine to the server

1. In the root path, use the command `ssh-keygen` to generate SSH keys
2. Create a file with the name `authorized_keys` in `.ssh` folder
3. Copy the public key here


## **3. Install tools on server: Docker, Git, etc**

- Create the given file `setup_server.sh`
- On terminal execute `chmod +x setup_server.sh`
- `./setup_server.sh` to run and setup tools

## **4. Git**

### 1. Grant access to each Git repo

### 2. Clone Git Repos
git clone git@github.com:mlalgotrading/realtime_data_mlcourse.git
git clone git@github.com:mlalgotrading/realtime_metatrader_mlcourse.git
git clone git@github.com:mlalgotrading/realtime_docker_mlcourse.git

## **5. Run Docker Containers**

- Copy or Create and then Configure .env
- Setup Docker Volume:
    1. docker volume create vscode-dev-1
    2. docker volume create reatime_data_ipython
    3. docker volume create reatime_data_cache
- Docker Compose Up
    `docker compose up --build`

## **6. Configure Broker on Web**

- Open the dockerized Metatrader on the web address `YOURIP:3001`
- Enter your username and password
- Increase historical data access

## **7. Apply Prediction Model**

- Copy your model to the `realtime_data` Container in the path *data/model*
- Attach to running container: `realtime_data`
- Configure your telegram bot variables. You need to set `TELEGRAM_CHAT_ID` and `TELEGRAM_CHAT_ID`/`TELEGRAM_BOT_TOKEN_SIGNAL` in the .env file. See [this tutorial](https://gist.github.com/nafiesl/4ad622f344cd1dc3bb1ecbe468ff9f8a).
- In the terminal run `tmux`
- In the tumx enter the command: `python -m jupyter nbconvert --execute --to notebook realtime.ipynb`
